```python
from sympy import *
from sympy.plotting import plot_parametric
from sympy.plotting import plot3d_parametric_line
import math

init_printing()
```

```python
t = symbols('t')
```

```python
x = cos(t)
y = sin(t)
plot_parametric(x, y, (t, 0, 2*pi))
```

```python
x = t + 0.5 * cos(10*t)
y = t + 0.5 * sin(10*t)
plot_parametric(x, y, (t, -2*pi, 2*pi))
```

```python
x = t -sin(t)
y = 1-cos(t)
plot_parametric(x, y, (t, -2*pi, 2*pi))
```

```python
x = 0.1*t*cos(t)
y = 0.1*t*sin(t)
plot_parametric(x, y, (t, 0, 10*pi))
```

```python
x = cos(t)
y = sin(t)
z = t

plot3d_parametric_line(x, y, z, (t, 0, 6*math.pi), size=(5,5))
```

```python
x = t*cos(t)
y = t*sin(t)
z = t

plot3d_parametric_line(x, y, z, (t, 0, 6*pi), size=(10,10))
```
