```python
from scipy.stats import qmc
sampler = qmc.Sobol(d=2, scramble=False)
sample = sampler.random_base2(m=7)
sample[:10]
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(sample[:,0], sample[:,1], c ="blue")
 
# To show the plot
plt.show()
```
