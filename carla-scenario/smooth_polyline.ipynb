```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([75, 0, 25, 100])
plt.plot(x, y)
```

```python
x_new = np.linspace(1, 4, 300)
a_BSpline = interpolate.make_interp_spline(x, y)
y_new = a_BSpline(x_new)

plt.plot(x_new, y_new)
```
