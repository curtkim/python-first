```python
import torch
import kornia
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
cuda = torch.device('cuda')
cpu = torch.device('cpu')
```

```python
img_bgr: np.ndarray = cv2.imread('./data/arturito.jpeg')  # H W C
```

```python
x_bgr: torch.tensor = kornia.image_to_tensor(img_bgr)  # 1 C H W
```

```python
x_bgr_cuda = x_bgr.to(device=cuda)
```

```python
x_rgb_cuda: torch.tensor = kornia.bgr_to_rgb(x_bgr_cuda)  # 1 C H W
```

```python
x_rgb = x_rgb_cuda.to(device=cpu)
```

```python
img_rgb: np.ndarray = kornia.tensor_to_image(x_rgb.byte())  # H W C
```

```python
fig, axs = plt.subplots(1, 2, figsize=(32, 16))
axs = axs.ravel()

axs[0].axis('off')
axs[0].imshow(img_bgr)

axs[1].axis('off')
axs[1].imshow(img_rgb)
```
