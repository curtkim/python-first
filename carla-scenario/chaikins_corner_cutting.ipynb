```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def chaikins_corner_cutting(coords, refinements=2):
    coords = np.array(coords)

    for _ in range(refinements):
        L = coords.repeat(2, axis=0)
        R = np.empty_like(L)
        R[0] = L[0]
        R[2::2] = L[1:-1:2]
        R[1:-1:2] = L[2::2]
        R[-1] = L[-1]
        coords = L * 0.75 + R * 0.25

    return coords
```

```python
coords = np.array([
    [1, 2, 3, 4],
    [2, 2, 1, 1]
])
plt.plot(coords[0], coords[1])
```

```python
coords2 = chaikins_corner_cutting(coords.T, 4).T
print(coords2.shape)
plt.plot(coords2[0], coords2[1])
```
