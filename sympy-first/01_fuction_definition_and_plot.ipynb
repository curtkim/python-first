```python
from sympy import *
init_printing()
```

```python
x, y, z = symbols('x y z')
```

```python
f = 2*x**3
f
```

```python
g = x**2 + x*y**2
g
```

## Function evaluation

```python
f.subs(x, 0)
```

```python
f.subs(x, 1)
```

```python
g.subs([(x,2), (y,3)])
```

## Plotting

```python
f=x**2
plot(f)
```

```python
f = x**2
g = x**3


p = plot(f, g, (x, -2, 2), show=False, legend=True, size=(10,5))
p[0].line_color = 'blue'
p[1].line_color = 'green'

p[0].label = 'f(x)'
p[1].label = 'g(x)'
p.show()
```
